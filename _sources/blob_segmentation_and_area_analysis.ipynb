# Blob Segmentation Notebook

This notebook demonstrates how to load a microscopy image, segment bright blobs using simple thresholding and connected component labeling, measure their areas, and compute the average area.

## Planned Steps

1. **Import required libraries**
2. **Load the image** (`data/blobs.tif`)
3. **Threshold the image** to obtain a binary mask of the bright blobs
4. **Label connected components** in the binary mask
5. **Measure object properties** (area) for each label
6. **Compute and print the average area** of the detected blobs

```python
import numpy as np
import pandas as pd
from skimage import io, filters, measure
import stackview
```

### Load the image

We read the TIFF stack located at `data/blobs.tif` and display it.

```python
# Load the image
image = io.imread('data/blobs.tif')

# Show the raw image
stackview.insight(image)
```

### Threshold the image

Apply Otsu's method to obtain a global threshold and create a binary mask of the bright blobs.

```python
# Compute Otsu threshold
threshold = filters.threshold_otsu(image)

# Create binary mask (blobs are brighter than the threshold)
binary_mask = image > threshold

# Show the binary mask
stackview.insight(binary_mask)
```

### Connected component labeling

Label each contiguous blob in the binary mask.

```python
# Label connected components (background = 0)
labels = measure.label(binary_mask, connectivity=1)

# Show the label image (different colors for different blobs)
stackview.insight(labels)
```

### Measure object areas

Extract the area (in pixels) of each labeled blob and store the results in a DataFrame.

```python
# Compute region properties (area)
props = measure.regionprops_table(labels, properties=['label', 'area'])

# Convert to pandas DataFrame for easy handling
df = pd.DataFrame(props)

# Display the first few rows
display(df.head())
```

### Compute and print the average area

```python
# Calculate average area
average_area = df['area'].mean()

# Print the result
print(f'Average blob area: {average_area:.2f} pixels')
```
